# 接雨水

给定 n 个非负整数表示每个宽度为 1 的柱子的高度图，计算按此排列的柱子，下雨之后能接多少雨水。

示例 1：

![接雨水示例图](https://assets.leetcode.cn/aliyun-lc-upload/uploads/2018/10/22/rainwatertrap.png)


输入：height = [0,1,0,2,1,0,1,3,2,1,2,1]

输出：6

解释：上面是由数组 [0,1,0,2,1,0,1,3,2,1,2,1] 表示的高度图，在这种情况下，可以接 6 个单位的雨水（蓝色部分表示雨水）。 

示例 2：

输入：height = [4,2,0,3,2,5]

输出：9
 
提示：

n == height.length

1 <= n <= 2 * 104

0 <= height[i] <= 10^5

## 求解思路

破局点在于转换视角：把整体拆解成局部。

核心解题思路：只看当前这一列。

与其去想一整个大水坑能装多少水，不如我们盯住某一个具体的柱子（例如索引为 i 的柱子），想想这个柱子头上能顶着多少水？

根据木桶效应，柱子 i 顶上的水位高度，取决于它左边最高的柱子和右边最高的柱子中，较矮的那一个，第 i 个柱子上的水深计算公式为：

$$水深 = \max(\min(\text{左边最高}, \text{右边最高}) - \text{当前柱子的高度}, 0)$$

note：如果计算出来小于0，说明当前柱子比两边都高，存不了水，记为0，也就是上式中max 0 的操作。

通过上述思路，我们只需要知道每个位置左右两侧柱子的最高值，就可以计算出每一个柱子头上能接多少雨水，相加起来即可得到最终的答案。

我们可以利用两个指针 left 和 right 从数组的两端向中间夹逼，一遍遍历即可实现，算法的时间复杂度为 n，空间复杂度为 1。



```python
class Solution:
    def trap(self, height: List[int]) -> int:
        # 一个柱子头上能顶多少水取决于两个因素：
        # 1. 左方向柱子的最高值l_max、右方向柱子的最高值r_max
        # 2. 柱子本身的高度 h
        # 计算公式：max(min(l_max, r_max) - h, 0)
        n = len(height)
        if n < 3:
            return 0
        ans = 0
        left, right = 1, n-2
        left_max, right_max = height[0], height[n-1]
        # 核心：左右指针哪边的最高的柱子更矮一些就能先确定哪边的水柱高度
        # 木桶效应，要两边都兜得住才行！左指针从左往右走，左指针之左的最高高度确定了，右指针从右往左走，右指针之右的最高高度也确定了。
        # 注意：这里的循环条件要包含等于，因为left、right是待处理的柱子，两指针相遇的时候也要处理那根柱子
        while left <= right:
            if left_max < right_max:
                ans += max(left_max - height[left], 0)
                left_max = max(height[left], left_max)
                left += 1
            else:
                ans += max(right_max - height[right], 0)
                right_max = max(height[right], right_max)
                right -= 1
        return ans
```
