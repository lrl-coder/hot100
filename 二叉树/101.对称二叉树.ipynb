# 101. 对称二叉树

给你一个二叉树的根节点 `root` ， 检查它是否轴对称。

## 示例 1

```text
输入：root = [1,2,2,3,4,4,3]
输出：true
```

## 示例 2

![示例2](https://pic.leetcode.cn/1698027008-nPFLbM-image.png)

```text
输入：root = [1,2,2,null,3,null,3]
输出：false
```

## 我的思路

最开始看见这个题，我的思路是中序遍历树，如果得到的是回文序列那么就是对称的，反之不对称。

这样看起来OK，能过很多用例，但是是不对的，例如所有的节点都是一个值，那么程序不管怎么运行结果都是轴对称。

### 为什么错？

中序遍历（左-根-右）确实能把树拍平，但它会丢失二叉树的结构信息（不知道谁是谁的左右孩子）。

例如：

```text
        1
       / \
      2   2
     /   /
    3   3
```

整棵树的中序遍历拼起来是：[3, 2, 1, 3, 2]，是一个完美的回文！但回到图上一看，这棵树根本不是对称的（右边的 3 必须长在右孩子的位置上才对称）。

## 使用递归实现

要判断一棵树是否对称，单看一个节点是不行的，我们需要比较的是两个节点：左子树和右子树。如果左子树和右子树是对称的，那么这棵树就是对称的。

用递归三要素求解：

1. 参数和返回值：参数是传入的左右两个节点，返回值是布尔值，判断这两个节点是否对称。
2. 终止条件
   1. 两个节点都为空：对称
   2. 一个空、一个不空：不对称
   3. 两个都不空：值不一样不对称
3. 单层逻辑：如果刚才的终止条件都没触发，说明这两个节点都不为空，且值相等。
那作为当前层，我的任务就是派小弟去查它们的下一代是不是镜像的：
   1. 外侧比较：左节点的左孩子，必须和右节点的右孩子对比。
   2. 内侧比较：左节点的右孩子，必须和右节点的左孩子对比。
   3. 只有外侧和内侧都对称，这俩节点才算真正的对称。

````python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def fn(self, left, right) -> bool:
        if not left and not right:
            return True
        if (not left and right) or (left and not right):
            return False
        if left.val != right.val:
            return False
        return self.fn(left.left, right.right) and self.fn(left.right, right.left)


    def isSymmetric(self, root: Optional[TreeNode]) -> bool:
        return self.fn(root.left, root.right)
````
