# 199. 二叉树的右视图

给定一个二叉树的 根节点 root，想象自己站在它的右侧，按照从顶部到底部的顺序，返回从右侧所能看到的节点值。

示例 1：

![example1](https://assets.leetcode.com/uploads/2024/11/24/tmpd5jn43fs-1.png)

> 输入：root = [1,2,3,null,5,null,4]
> 
> 输出：[1,3,4]

示例 2：

![example2](https://assets.leetcode.com/uploads/2024/11/24/tmpkpe40xeh-1.png)

> 输入：root = [1,2,3,4,null,null,null,5]
> 
> 输出：[1,3,4,5]

## 思路

右视图的本质就是每一层最右边的节点。

利用层次遍历将树分层，把每一层的最右端加入到答案中。

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def rightSideView(self, root: Optional[TreeNode]) -> List[int]:
        # 用层次遍历，存储每一层的最右节点
        import collections
        q = deque()
        ans = []

        if root:
            q.append(root)

        while q:
            # 当前层的最右侧节点，一定是在队尾
            ans.append(q[-1].val)
            n = len(q)
            for i in range(n):
                t = q.popleft()
                if t.left:
                    q.append(t.left)
                if t.right:
                    q.append(t.right)
            
        return ans
        
```
