# 230. 二叉搜索树中第 K 小的元素

给定一个二叉搜索树的根节点 root ，和一个整数 k ，请你设计一个算法查找其中第 k 小的元素（k 从 1 开始计数）。

示例 1：

![example1](https://assets.leetcode.com/uploads/2021/01/28/kthtree1.jpg)

> 输入：root = [3,1,4,null,2], k = 1
> 
> 输出：1

示例 2：

![example1](https://assets.leetcode.com/uploads/2021/01/28/kthtree2.jpg)

> 输入：root = [5,3,6,2,4,null,null,1], k = 3
> 
> 输出：3

## 思路

中序遍历二叉搜索树得到的序列是递增序列。

利用中序遍历（左-根-右）直接按从小到大的顺序访问节点，当访问到第 k 个节点时，它就是第 k 小的元素。

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def kthSmallest(self, root: Optional[TreeNode], k: int) -> int:
        # 中序遍历，找到第k个遍历的元素即可
        ans = None
        def mid_travel(root):
            nonlocal ans, k
            if not root or not k:
                return None
            mid_travel(root.left)
            k -= 1
            if k == 0:
                ans = root.val
                return
            mid_travel(root.right)
        
        mid_travel(root)
        return ans

        
```
