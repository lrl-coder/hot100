# 105. 从前序与中序遍历序列构造二叉树

给定两个整数数组 preorder 和 inorder ，其中 preorder 是二叉树的先序遍历， inorder 是同一棵树的中序遍历，请构造二叉树并返回其根节点。

示例 1:

![1](https://assets.leetcode.com/uploads/2021/02/19/tree.jpg)

输入: preorder = [3,9,20,15,7], inorder = [9,3,15,20,7]

输出: [3,9,20,null,null,15,7]

提示:

- 1 <= preorder.length <= 3000
- inorder.length == preorder.length
- -3000 <= preorder[i], inorder[i] <= 3000
- preorder 和 inorder 均 无重复 元素
- inorder 均出现在 preorder
- preorder 保证 为二叉树的前序遍历序列
- inorder 保证 为二叉树的中序遍历序列

## 解题思路

这是一道公式题，要根据前序遍历（Preorder）和中序遍历（Inorder）的特点来破题。

前序遍历：根左右，根在最前面；中序遍历：左根右，根在中间。

由此我们就可以根据前序遍历的根不断划分中序序列，把中序序列分为左子树和右子树，然后递归求解。

**理论化阐述**：

- 前序遍历（根-左-右）： 数组的第一个元素永远是当前树（或子树）的根节点。
- 中序遍历（左-根-右）： 找到根节点在中序数组中的位置后，根节点会将中序数组一分为二：左边的所有元素构成左子树，右边的所有元素构成右子树。

可以将问题分解为以下几个步骤：

1. 找根节点：从前序遍历中取出第一个节点的值；
2. 切分中序序列：在中序遍历中找到该节点的索引(元素不重复) mid，mid左侧的序列是左子树的中序遍历，右侧是右子树的中序遍历。
3. 计算子树长度：通过中序遍历中左子树的元素个数(left_size=mid-in_start)，我们可以推断出：在前序遍历中，紧接着根节点的后面left_size个元素就是左子树的前序遍历，剩下的则是右子树的前序遍历。
4. 递归构建：将划分好的前序和中序序列分别递归构建左右子树，并将它们挂载到当前的根节点上。

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def buildTree(self, preorder: List[int], inorder: List[int]) -> Optional[TreeNode]:
        # 函数的参数：先序遍历、中序遍历序列
        # 返回值：创建的根节点
        def fn(pre_start, pre_end, in_start, in_end):
            # 递归结束条件
            if pre_start > pre_end:
                return None
            # 这一层要做的事：划分序列
            # 根节点的值
            root_val = preorder[pre_start]
            in_mid = inorder.index(root_val)
            # 左子树的元素个数
            len_left = in_mid - in_start
            # 右子树的元素个数
            len_right = in_end - in_mid
            # 左子树的根
            root_left = fn(pre_start+1, pre_start+len_left, in_start, in_mid-1)
            # 右子树的根
            root_right = fn(pre_start+len_left+1, pre_end, in_mid+1, in_end)
            root = TreeNode(val=root_val, left=root_left, right=root_right)
            return root
        return fn(0, len(preorder)-1, 0, len(inorder)-1)
```
