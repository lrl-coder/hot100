
# 108. 将有序数组转换为二叉搜索树

给你一个整数数组 `nums` ，其中元素已经按 **升序** 排列，请你将其转换为一棵 平衡 二叉搜索树。

```text
输入：nums = [-10,-3,0,5,9]
输出：[0,-3,9,-10,null,5]

````

## 示例 2

![示例2](https://assets.leetcode.com/uploads/2021/02/18/btree.jpg)

```text
输入：nums = [1,3]
输出：[3,1]
解释：[1,null,3] 和 [3,1] 都是高度平衡二叉搜索树。
```





## 求解思路

要解决这个问题，我们需要抓住两个关键点：

1. 二叉搜索树（BST）的性质：左子树的所有节点值小于根节点，右子树的所有节点值大于根节点。题目中给定的数组是升序排序的，符合BST中序遍历的结果。
2. 高度平衡的性质：平衡二叉树要求左右两个子树的高度差的绝对值不超过1，为了做到这一点，我们需要让左右子树分配到的节点数量尽可能一样多。

**这道题的破局点就在于“升序数组”和“高度平衡”的结合。**

假设我们随便拿数组的第一个元素作为树的根节点（比如数组 [-10, -3, 0, 5, 9]，我们选 -10 作根），那么剩下的所有数都比它大，只能全部塞到它的右子树里。结果就是这棵树完全向右倾斜，高度极其不平衡。那要怎么样才能平衡呢？

最直观的办法就是从正中间切开：
- 把数组正中间的元素作为根节点。
- 因为数组是升序的，所以中间元素左边的数一定都比它小，刚好可以打包成左子树。
- 中间元素右边的数一定都比它大，刚好打包成右子树。
- 因为是从正中间切的，左右两边的数字个数要么相等，要么最多差 1 个，这样自然就保证了“高度平衡”。

如此，这个问题就变成了经典的分治法：把一个大问题（整个数组转成树），拆成了两个一模一样的小问题（左半边数组转成左子树，右半边数组转成右子树）。

递归三件套：

1. 函数的参数：数组(可用左右指针表示)，返回值：创建的树的根节点
2. 终止条件：数组元素为空
3. 这一层要做的事：创建根节点，调用函数得到左子树右子树，指针指向左子树右子树。

`````python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def sortedArrayToBST(self, nums: List[int]) -> Optional[TreeNode]:
        def bst(left, right):
            if left > right:
                return None
            
            mid = (left + right) // 2            
            left = bst(left, mid-1)
            right = bst(mid+1, right)
            node = TreeNode(val=nums[mid], left=left, right=right)
            return node
        
        return bst(0, len(nums)-1)
`````
