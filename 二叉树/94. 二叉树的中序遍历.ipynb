# 94. 二叉树的中序遍历

给定一个二叉树的根节点 root ，返回 它的 中序 遍历 。

## 思路

解这道题的思路很清晰：左根右，按照这个顺序去遍历就好。

但是在具体写代码时，很容易翻车，总结一下错误原因：

很多时候我们写错递归，是因为大脑在试图模拟整个调用栈的运行过程，想把树的每一层、每一个分支怎么进去、怎么出来全想清楚，结果脑内存瞬间溢出，逻辑就乱了。

但是事实上，你根本不需要在脑子里把前向传播和反向传播的千万个节点死磕到底。你只需要定义好当前这个单一节点的局部逻辑，剩下的引擎（系统调用栈）会自动帮你完成流转。

1. 建立“黑盒信仰”：这是写递归最重要的一步。永远不要在脑海里往下追溯超过一层。
2. 聚焦“当前节点的职责”：对于当前的节点，我需要做什么？
3. 严格落实“递归三要素”模板

递归三要素模板：

一、 确定函数的参数和返回值（定义黑盒）

在动手写代码前，先明确这个函数“需要接收什么材料”以及“必须向上层交付什么结果”。

核心思维： 明确这个函数到底是用来干嘛的，把它当成一个别人已经写好的、绝对可靠的系统 API。
例子： 假设要计算 $n$ 的阶乘，参数就是一个整数 $n$，返回值就是算好的阶乘结果：def factorial(n: int) -> int:。

二、 确定终止条件（寻找出口）这是递归的“刹车踏板”。

计算机的内存是有限的，如果不告诉它什么时候停下，函数就会无限往下调用自己，直到程序崩溃（也就是常见的 Stack Overflow 栈溢出）。核心思维： 当问题被拆解到最小、最简单，不需要再往下递推就能直接给出答案的时候，是什么状态？例子： 在算阶乘时，当 $n = 1$ 时，我知道结果就是 $1$，不需要再往下算了，这就是出口：if n == 1: return 1。

三、 确定单层递归逻辑（当前节点职责）

这是最关键的一步。千万不要去想全局，只关注当前这一层我要做什么工作。核心思维： 当前这一层，我该怎么利用“下一层返回的局部结果”，来拼凑出“当前层需要的最终结果”？例子： 我要算 $n!$，我不想知道底下的细节，我只需要黑盒帮我算出 $(n-1)!$，然后我拿当前的 $n$ 乘以那个结果就行了：return n * factorial(n - 1)。

```python
# 要素一：参数是 n，返回值是 int
def factorial(n: int) -> int:
    
    # 要素二：终止条件
    if n == 1:
        return 1
        
    # 要素三：单层递归逻辑
    # 相信 factorial(n - 1) 一定能正确返回 n-1 的阶乘，我只负责乘上 n
    return n * factorial(n - 1)
```

````python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def inorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
        # 1. 函数的参数：当前处理的节点，返回值：中序遍历序列
        # 2. 这一层要做的事，返回这一节点的序列：左序列 + 当前节点的值 + 右序列
        # 3. 递归终止条件：节点为空，返回[]

        if not root:
            return []

        left = self.inorderTraversal(root.left)
        right = self.inorderTraversal(root.right)

        return left + [root.val] + right
````
