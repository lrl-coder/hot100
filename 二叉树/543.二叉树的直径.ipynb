# 543. 二叉树的直径

给你一棵二叉树的根节点，返回该树的 **直径** 。

二叉树的 **直径** 是指树中任意两个节点之间最长路径的 **长度** 。这条路径可能经过也可能不经过根节点 `root` 。

两节点之间路径的 **长度** 由它们之间边数表示。

## 示例 1

![示例1](https://assets.leetcode.com/uploads/2021/03/06/diamtree.jpg)

```text
输入：root = [1,2,3,4,5]
输出：3
解释：3 ，取路径 [4,2,1,3] 或 [5,2,1,3] 的长度。

## 求解思路

以某个节点为根节点的直径等于左子树的深度+右子树深度，所有节点中的最大值即为题目中求的树的直径。

````python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    ans = 0

    def depth(self, root):
        if not root.left and not root.right:
            return 0
        left, right = 0, 0
        if root.left:
            left = self.depth(root.left) + 1
        if root.right:
            right = self.depth(root.right) + 1
        current_depth = left + right
        self.ans = max(self.ans, current_depth)
        return max(left, right)
        

    def diameterOfBinaryTree(self, root: Optional[TreeNode]) -> int:
        self.depth(root)
        return self.ans
        
````

### 优化

上面的代码能够ac，但是看起来还是有点冗余，原因在于微操太多，还是不敢完全信任递归黑盒。

核心还是思维问题，递归算的是深度，只是通过ans记录直径！


````python
class Solution:
    def diameterOfBinaryTree(self, root: Optional[TreeNode]) -> int:
        ans = 0
        
        # 定义黑盒：计算并返回以 node 为根的树的最大深度
        def depth(node: Optional[TreeNode]) -> int:
            nonlocal ans  # 允许在内部函数中修改外部变量 ans
            
            # --- 1. 终止条件 ---
            if not node:
                return 0
            
            # --- 2. 扔进黑盒 ---
            # 完全信任黑盒，不要去管 left 是不是 None，直接扔进去！
            left_depth = depth(node.left)
            right_depth = depth(node.right)
            
            # --- 3. 单层逻辑 ---
            # 职责A：计算如果路径穿过当前 node，能有多长？并更新全局最大值
            ans = max(ans, left_depth + right_depth)
            
            # 职责B：向上级汇报我这棵树的最大深度
            return max(left_depth, right_depth) + 1
            
        # 启动计算
        depth(root)
        return ans
````
