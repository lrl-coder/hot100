# 124. 二叉树中的最大路径和

和543很像！

二叉树中的 路径 被定义为一条节点序列，序列中每对相邻节点之间都存在一条边。同一个节点在一条路径序列中 至多出现一次 。该路径 至少包含一个 节点，且不一定经过根节点。

路径和 是路径中各节点值的总和。

给你一个二叉树的根节点 root ，返回其 最大路径和 。

示例1：

![1](https://assets.leetcode.com/uploads/2020/10/13/exx1.jpg)

输入：root = [1,2,3]

输出：6

解释：最优路径是 2 -> 1 -> 3 ，路径和为 2 + 1 + 3 = 6

示例2：

![2](https://assets.leetcode.com/uploads/2020/10/13/exx2.jpg)

输入：root = [-10,9,20,null,null,15,7]

输出：42

解释：最优路径是 15 -> 20 -> 7 ，路径和为 15 + 20 + 7 = 42

## 思路

采用深度优先搜索，搜索的目标是以当前节点为根的子树且经过当前节点的最大路径。在搜索的过程中记录最大答案。

**特别注意**：寻找最大答案时，我们可以把左右子树的路径和加上（如果大于零的话）；但是如果要传递给上一层，只能选择加左子树、加右子树、什么都不加（左右子树都小于0），否则的话就串不起来了。

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    
    def maxPathSum(self, root: Optional[TreeNode]) -> int:
        ans = float('-inf')
        # 递归函数定义：以root为根节点的子树且经过root的最大路径和
        def fn(root):
            nonlocal ans
            if not root:
                return 0

            curr_sum = root.val
            left = fn(root.left)
            right = fn(root.right)
            if left > 0:
                curr_sum += left
            if right > 0:
                curr_sum += right
            # 结果取最大值
            ans = max(ans, curr_sum)
            # 上层如果要用的话只能走一边
            if left < 0 and right < 0:
                return root.val
            return root.val + left if left > right else root.val + right
        fn(root)
        return ans
            
        
```
