# 旋转图像

给定一个 *n* × *n* 的二维矩阵 `matrix` 表示一个图像。请你将图像顺时针旋转 90 度。

你必须在**原地**旋转图像，这意味着你需要直接修改输入的二维矩阵。**请不要**使用另一个矩阵来旋转图像。

## 提示

* `n == matrix.length == matrix[i].length`
* `1 <= n <= 20`
* `-1000 <= matrix[i][j] <= 1000`


## 观察规律

模拟矩阵的旋转，我们可以发现以下规律：

- 第 i 行元素旋转到第 n-1-i 列
- 第 j 列元素旋转到第 j 行

也就是 a[i][j] 旋转之后的位置为 a[j][n-1-i]

![旋转的规律](https://pic.leetcode.cn/1638557961-AVzCQb-ccw-01-07.001.png)

由此我们可以快速得到一个简单但不符合题目要求的写法，创建一个辅助数组，然后利用公式逐元素赋值即可，下面先给出代码，后续再优化：


````python
class Solution:
    def rotate(self, matrix: List[List[int]]) -> None:
        """
        Do not return anything, modify matrix in-place instead.
        """
        # a[i][j] -> a[j][n-1-i]
        import copy
        tmp = deepcopy(matrix)
        n = len(matrix)
        for i in range(n):
            for j in range(n):
                matrix[j][n-1-i] = tmp[i][j]       
````

## 优化

观察旋转位置公式，a[i][j] -> a[j][n-1-i]，我们可以得到一个惊人的发现！

1. 沿对角线翻转：a[i][j] -> a[j][i]
2. 每一行水平翻转：a[i][j] -> a[i][n-1-i]

因此，旋转公式就等价于先沿对角线翻转，然后每一行再水平翻转。

````python
class Solution:

    def rotate(self, matrix: List[List[int]]) -> None:
        """
        Do not return anything, modify matrix in-place instead.
        """
        n = len(matrix)
        for i in range(n):
            for j in range(i):
                matrix[i][j], matrix[j][i] = matrix[j][i], matrix[i][j]
                
        for i in range(n):
            matrix[i].reverse()
````
