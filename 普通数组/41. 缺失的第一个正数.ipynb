41. 缺失的第一个正数

给你一个未排序的整数数组 nums ，请你找出其中没有出现的最小的正整数。

请你实现时间复杂度为 O(n) 并且只使用常数级别额外空间的解决方案。
 

示例 1：

输入：nums = [1,2,0]

输出：3

解释：范围 [1,2] 中的数字都在数组中。

示例 2：

输入：nums = [3,4,-1,1]

输出：2

解释：1 在数组中，但 2 没有。

示例 3：

输入：nums = [7,8,9,11,12]

输出：1

解释：最小的正数 1 没有出现。
 
提示：

1 <= nums.length <= 10^5

-2^31 <= nums[i] <= 2^31 - 1

## 哈希实现（空间复杂度超了）

一开始看见这个题本来打算通过看元素左右的正数值有没有出现，来计算答案。但是行不通，因为可能会出现断层的现象，如 [8, 9]。

翻阅题解后看见了一种哈希解法，虽然不符合题目中对空间复杂度的要求，但是思路很巧妙，下面给出思路及源码：

**核心思路**：一个萝卜一个坑对于一个长度为 $N$ 的数组，其中缺失的最小正整数一定在 $[1, N+1]$ 的范围内。因为无非就两种情况：
- 情况 1：如果数组是 [1, 2, 3, ..., N]，缺失的最小正整数就是 $N+1$。
- 情况 2：如果数组包含负数、大于 $N$ 的数，或者有重复的数，那么缺失的最小正整数必然在 $1$ 到 $N$ 之间。


```python
class Solution:
    def firstMissingPositive(self, nums: List[int]) -> int:
        # 核心思路：一个萝卜一个坑
        # 对于一个长度为 N 的数组，其中缺失的最小正整数一定在 [1, N+1] 的范围内。
        # 正确性证明：对于一个长度为 N 的数组只有N个元素，无非两种情况
        # 1. 1~N之间的数都出现在了数组中，那么答案就是N+1
        # 2. 1~N之间有某个数没有出现在数组中，那么答案就是该数
        # 通过上面的分析可以看出，这又是一个查找值的过程，所以可以用哈希实现
        # 因为这里只是判断有没有出现，不需要记录索引等额外信息，所以用 set 即可
        st = set(nums)
        n = len(nums)
        for i in range(1, n+2):
            if i not in st:
                return i
```

## 优化

我们可以把原数组本身当做一个 哈希表 来用，我们期望达到的目标状态是：数字 x 被放在索引 x - 1 的位置上（例如数组1放在索引0），放置好了之后，循环遍历数组：
- 如果哪一个位置没有摆放正确的数字，那么就表明缺少该数字，答案就是该数字；
- 如果所有位置都摆放了正确的元素，那么答案就是 n + 1

难点在于**如何将数字摆放到正确的位置**：

如果数字x不在正确的位置上，我们需要看他的目标位置上的元素：

- 与 x 一样，那就不用管了，x是重复的元素，数字x已经摆好了位置；
- 与 x 不一样，与 x 交换位置，并循环调整

下面给出代码：


```python
class Solution:
    def firstMissingPositive(self, nums: List[int]) -> int:
        # 原地哈希，让nums依次存储1~N数字
        n = len(nums)
        for i, num in enumerate(nums):
            while 0 < nums[i] <= n and nums[nums[i]-1] != nums[i]:
                nums[nums[i]-1], nums[i] = nums[i], nums[nums[i]-1]
        for i, num in enumerate(nums):
            if num-1 != i:
                return i+1
        return n+1
```
