# 最大子数组和

给你一个整数数组 nums ，请你找出一个具有最大和的连续子数组（子数组最少包含一个元素），返回其最大和。

子数组是数组中的一个连续部分。 

示例 1：

输入：nums = [-2,1,-3,4,-1,2,1,-5,4]

输出：6

解释：连续子数组 [4,-1,2,1] 的和最大，为 6 。

示例 2：

输入：nums = [1]

输出：1

示例 3：

输入：nums = [5,4,-1,7,8]

输出：23
 
提示：

1 <= nums.length <= 105

-104 <= nums[i] <= 104 

进阶：如果你已经实现复杂度为 O(n) 的解法，尝试使用更为精妙的 分治法 求解。


这个题我记得本科上的算法设计与分析有讲过，当时是用分治法做的，分成左边、右边、跨越左右两边的，最终的结果就是取这三者的最大值，用递归实现。

因为用递归实现很麻烦，我就想着尝试用双指针求解，成功通过，算法的时间复杂度为 O(n)。

解题的思路：

首先定义左右指针，指向子数组的左右两端，然后循环移动右指针不断遍历数组。用一个 tmp 变量计算子数组的和。

核心思想是及时止损，当子数组的和小于 0 或者左指针指向的元素小于 0 时，不断抛弃子数组最左端的元素，并更新 tmp 的值。

下面给出源代码：

```python
class Solution:
    def maxSubArray(self, nums: List[int]) -> int:
        n = len(nums)
        ans = nums[0]
        tmp = 0
        # 定义左右指针，指向子数组的左右两端
        left, right = 0, 0
        # 核心思想：用右指针不断遍历数组，及时止损
        # 当子数组的和小于0或者左指针指向的元素小于0时循环抛弃左端的元素，直到和不小于0为止
        while right < n:
            tmp += nums[right]
            while (tmp < 0 or nums[left] < 0) and left < right:
                tmp -= nums[left]
                left += 1
            ans = max(ans, tmp)
            right += 1
        return ans
```

## 动态规划

上面的双指针虽然能够成功解答，但是设计逻辑边界问题，很容易出错，下面给出动态规划的解法，更容里理解：

dp 数组：dp[i] 代表以第 i 个元素为结束的最大子数组和，那么：

dp[i] = dp[i - 1] + nums[i], if dp[i - 1] > 0;

dp[i] = nums[i], if dp[i - 1] <= 0;

下面给出代码：



```python
class Solution:
    def maxSubArray(self, nums: List[int]) -> int:
        n = len(nums)
        dp = [0 for _ in range(n)]
        dp[0] = nums[0]
        for i in range(1, n):
            if dp[i - 1] < 0:
                dp[i] = nums[i]
            else:
                dp[i] = dp[i-1] + nums[i]
        return max(dp)
```
