238. 除了自身以外数组的乘积

给你一个整数数组 nums，返回 数组 answer ，其中 answer[i] 等于 nums 中除了 nums[i] 之外其余各元素的乘积 。

题目数据 保证 数组 nums之中任意元素的全部前缀元素和后缀的乘积都在  32 位 整数范围内。

请 不要使用除法，且在 O(n) 时间复杂度内完成此题。

示例 1:

输入: nums = [1,2,3,4]

输出: [24,12,8,6]

示例 2:

输入: nums = [-1,1,0,-3,3]

输出: [0,0,9,0,0] 

提示：


2 <= nums.length <= 105

-30 <= nums[i] <= 30

输入 保证 数组 answer[i] 在  32 位 整数范围内
 
进阶：你可以在 O(1) 的额外空间复杂度内完成这个题目吗？（ 出于对空间复杂度分析的目的，输出数组 不被视为 额外空间。）

```python
class Solution:
    def productExceptSelf(self, nums: List[int]) -> List[int]:
        # 两次循环
        # 找左边的乘积
        # 找右边的乘积
        # 左乘右就是答案
        n = len(nums)
        left = [1]*n
        right = [1]*n
        for i in range(1, len(nums)):
            left[i] = left[i-1]*nums[i-1]
        for i in range(len(nums)-2, -1, -1):
            right[i] = right[i+1]*nums[i+1]
        ans = []
        for i in range(len(nums)):
            ans.append(left[i]*right[i])
        return ans
```

## 优化：不开额外 left、right 数组

第一遍正向遍历，直接把左侧乘积存进 ans 里。

第二遍反向遍历时，不需要再建一个 right 数组，而是只用一个单独的变量 R（代表当前元素右侧的所有数字乘积）。一边往左走，一边更新 R，同时把 R 乘到 ans 对应的位置里。

```python
class Solution:
    def productExceptSelf(self, nums: List[int]) -> List[int]:
        n = len(nums)
        ans = [1] * n
        
        # 第一步：计算左侧乘积，直接存入 ans
        # 此时 ans[i] 里装的就是 nums[i] 左边所有元素的乘积
        for i in range(1, n):
            ans[i] = ans[i-1] * nums[i-1]
            
        # 第二步：计算右侧乘积，并直接乘到 ans 里
        R = 1 # R 用来动态记录右侧元素的累积乘积
        for i in range(n-1, -1, -1):
            ans[i] = ans[i] * R  # 左边乘积 * 右边乘积
            R *= nums[i]         # 维护 R，让它包含当前元素，供下一个元素(左边的元素)使用
            
        return ans
```
