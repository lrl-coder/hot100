189. 轮转数组

给定一个整数数组 nums，将数组中的元素向右轮转 k 个位置，其中 k 是非负数。

示例 1:

输入: nums = [1,2,3,4,5,6,7], k = 3

输出: [5,6,7,1,2,3,4]

解释:

向右轮转 1 步: [7,1,2,3,4,5,6]

向右轮转 2 步: [6,7,1,2,3,4,5]

向右轮转 3 步: [5,6,7,1,2,3,4]

示例 2:

输入：nums = [-1,-100,3,99], k = 2

输出：[3,99,-1,-100]

解释: 

向右轮转 1 步: [99,-1,-100,3]

向右轮转 2 步: [3,99,-1,-100] 

提示：

1 <= nums.length <= 10^5

-2^31 <= nums[i] <= 2^31 - 1

0 <= k <= 10^5 

进阶：

尽可能想出更多的解决方案，至少有 三种 不同的方法可以解决这个问题。

你可以使用空间复杂度为 O(1) 的 原地 算法解决这个问题吗？


## 思路1：利用python的冒号语法

后 n-k 个元素挪到最前，其余元素拼到后面。


```python
class Solution:
    def rotate(self, nums: List[int], k: int) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        n = len(nums)
        k = k % n
        nums[:] = nums[n-k:] + nums[:n-k]
```

## 炫技：三次翻转

1. 反转整个数组
2. 反转前 k 个元素
3. 反转剩余的 n-k 个元素

假设 nums = [1, 2, 3, 4, 5, 6, 7], k = 3：

第一步：反转整个数组

原数组：[1, 2, 3, 4, 5, 6, 7]  (也就是 A + B)

整体反转后：[7, 6, 5, 4, 3, 2, 1]

发生了什么？ 我们成功地把 B 块丢到了最前面，把 A 块丢到了最后面！它变成了 反转的B + 反转的A。

现在的缺陷： 虽然两块大拼图的位置对调了，但是它们内部的数字顺序是倒过来的。

第二步：反转前 k 个元素（即现在的 B 块）

当前数组：[7, 6, 5, 4, 3, 2, 1]

前 3 个反转：[5, 6, 7, 4, 3, 2, 1]

发生了什么？ B 块 [7, 6, 5] 被重新理顺成了 [5, 6, 7]。前半部分的大功告成！

第三步：反转剩下的 n-k 个元素（即现在的 A 块）

当前数组：[5, 6, 7, 4, 3, 2, 1]

后 4 个反转：[5, 6, 7, 1, 2, 3, 4]

发生了什么？ A 块 [4, 3, 2, 1] 被重新理顺成了 [1, 2, 3, 4]。后半部分也大功告成！


class Solution:
    def rotate(self, nums: List[int], k: int) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        k = k % len(nums)
        nums.reverse()
        nums[:k] = reversed(nums[:k])
        nums[k:] = reversed(nums[k:])
