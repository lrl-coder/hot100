# 22. 括号生成

数字 n 代表生成括号的对数，请你设计一个函数，用于能够生成所有可能的并且 有效的 括号组合。

 

示例 1：
```text
输入：n = 3
输出：["((()))","(()())","(())()","()(())","()()()"]
```
示例 2：
```text
输入：n = 1
输出：["()"]
```

## 解题思路

这是一个带限制条件的回溯。

这道题的选择其实非常单纯：在任何一个位置，就两种选择，你要么放左括号 (，要么放右括号 )。

限制：
- 限制 1（左括号）： 只要左括号用的数量还没达到 $n$ 个，我们随时可以往里加左括号 (。
- 限制 2（右括号）： 只有当已经使用的左括号数量 > 已经使用的右括号数量时，我们才能放右括号 )。这是因为右括号必须得有前面的左括号来跟它配对。如果你一共才放了 1 个左括号，却想放 2 个右括号，那必定是非法的。

````python
class Solution:
    def generateParenthesis(self, n: int) -> List[str]:
        ans = []

        # 当前字符串，左括号个数，右括号个数
        def fn(s, left_cnt, right_cnt):
            print(s)
            if left_cnt+right_cnt == 2*n:
                ans.append(s)
                return
            # 放左括号
            if left_cnt < n:
                s += '('
                left_cnt += 1
                fn(s, left_cnt, right_cnt)
                # 回溯
                s = s[:-1]
                left_cnt -= 1
            # 放右括号
            if right_cnt < left_cnt:
                s += ')'
                right_cnt += 1
                fn(s, left_cnt, right_cnt)
        
        fn('', 0, 0)

        return ans
````

优雅版代码：

Python 中的字符串是不可变对象（Immutable），对字符串进行 + 拼接操作时，实际上是生成了一个全新的字符串。

利用这个特性，我们可以直接把字符串作为路径传给下一层递归。因为传下去的是新对象，当前层的字符串根本没有被修改，所以我们不需要手动执行撤销选择的操作（不需要删除最后一个字母），代码会变得极其简短：

````python
class Solution:
    def generateParenthesis(self, n: int) -> List[str]:
        ans = []

        # 当前字符串，左括号个数，右括号个数
        def fn(s, left_cnt, right_cnt):
            print(s)
            if left_cnt+right_cnt == 2*n:
                ans.append(s)
                return
            # 放左括号
            if left_cnt < n:
                fn(s+'(', left_cnt+1, right_cnt)
            # 放右括号
            if right_cnt < left_cnt:
                fn(s+')', left_cnt, right_cnt+1)
        
        fn('', 0, 0)

        return ans
````
