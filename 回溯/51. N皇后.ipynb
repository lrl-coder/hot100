# 51. N 皇后

按照国际象棋的规则，皇后可以攻击与之处在同一行或同一列或同一斜线上的棋子。

n 皇后问题 研究的是如何将 n 个皇后放置在 n×n 的棋盘上，并且使皇后彼此之间不能相互攻击。

给你一个整数 n ，返回所有不同的 n 皇后问题 的解决方案。

每一种解法包含一个不同的 n 皇后问题 的棋子放置方案，该方案中 'Q' 和 '.' 分别代表了皇后和空位。

示例 1：
![1](https://assets.leetcode.com/uploads/2020/11/13/queens.jpg)
```text
输入：n = 4
输出：[[".Q..","...Q","Q...","..Q."],["..Q.","Q...","...Q",".Q.."]]
解释：如上图所示，4 皇后问题存在两个不同的解法。
```
示例 2：
```text
输入：n = 1
输出：[["Q"]]
```

## 解题思路

**核心的破题视角**：化二维为一维，按行推进，严密封锁。

由于皇后的攻击规则是“同行、同列、同斜线”，这天然赋予了我们一个最强大的剪枝条件：每一行绝对只能放一个皇后。

因此，我们的探索视角不再是“遍历棋盘”，而是：“我现在来到了第 $r$ 行，我只需要在这一行的 $N$ 个格子里挑一个安全的位置放皇后，放完立刻去下一行。” 这样，我们就彻底干掉了“行”的冲突，只需要去检查列和斜线。

**状态封锁**：
1. 列封锁 (col)：
创建一个长度为 $N$ 的数组。如果第 $c$ 列已经放了皇后，那么 col[c] = True。
2. 主对角线封锁 (diag1，即 \ 方向)：
观察棋盘的坐标规律，在同一条主对角线上的所有格子，它们的 行号减去列号的值是一个常数（即 $r - c = \text{const}$）。
为了防止出现负数索引，我们统一加上一个偏移量 $N - 1$。所以用 diag1[r - c + N - 1] = True 来封锁这条线。
3. 副对角线封锁 (diag2，即 / 方向)：
同理，在同一条副对角线上的格子，它们的 行号加上列号的值是一个常数（即 $r + c = \text{const}$）。
所以直接用 diag2[r + c] = True 来封锁。

> 这里的常量const可以当做下标查找

````python
class Solution:
    def solveNQueens(self, n: int) -> List[List[str]]:
        # 判断列、主对角线、反对角线是否可以放置皇后
        col = [True]*n
        diag_1 = [True]*(2*n-1)
        diag_2 = [True]*(2*n-1)

        ans = []
        path = [['.']*n for _ in range(n)]

        # 摆放第k行的皇后
        def backtrack(k):
            if k == n:
                ans.append([''.join(x) for x in path])
                return
            for i in range(n):
                if col[i] and diag_1[k-i+n-1] and diag_2[k+i]:
                    # 选择
                    col[i], diag_1[k-i+n-1], diag_2[k+i] = False, False, False
                    path[k][i] = 'Q'
                    # 递归
                    backtrack(k+1)
                    # 回溯
                    col[i], diag_1[k-i+n-1], diag_2[k+i] = True, True, True
                    path[k][i] = '.'
            
        backtrack(0)

        return ans
````
