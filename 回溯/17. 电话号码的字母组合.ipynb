# 17. 电话号码的字母组合

给定一个仅包含数字 2-9 的字符串，返回所有它能表示的字母组合。答案可以按 任意顺序 返回。

给出数字到字母的映射如下（与电话按键相同）。注意 1 不对应任何字母。

![1](https://pic.leetcode.cn/1752723054-mfIHZs-image.png)

示例 1：
```text
输入：digits = "23"
输出：["ad","ae","af","bd","be","bf","cd","ce","cf"]
```
示例 2：
```text
输入：digits = "2"
输出：["a","b","c"]
```

````python
class Solution:
    def letterCombinations(self, digits: str) -> List[str]:
        ans = []
        num2char = {
            '2': 'abc',
            '3': 'def',
            '4': 'ghi',
            '5': 'jkl',
            '6': 'mno',
            '7': 'pqrs',
            '8': 'tuv',
            '9': 'wxyz'
        }

        def fn(index, current_chars):
            if index == len(digits):
                ans.append(current_chars)
                return
            chars = num2char[digits[index]]
            for c in chars:
                tmp = current_chars
                current_chars = current_chars + c
                # 递归
                fn(index+1, current_chars)
                # 回溯
                current_chars = tmp
        
        fn(0, '')

        return ans
````
