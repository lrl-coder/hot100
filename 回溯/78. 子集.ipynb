# 78. 子集

给你一个整数数组 nums ，数组中的元素 互不相同 。返回该数组所有可能的子集（幂集）。

解集 不能 包含重复的子集。你可以按 任意顺序 返回解集。

 

示例 1：
```text
输入：nums = [1,2,3]
输出：[[],[1],[2],[1,2],[3],[1,3],[2,3],[1,2,3]]
```
示例 2：
```text
输入：nums = [0]
输出：[[],[0]]
```

## 解题思路

循环遍历数组的所有元素，每一个元素都有两种旋转：用/不用。

````python
class Solution:
    def subsets(self, nums: List[int]) -> List[List[int]]:
        ans = []
        # 用或者不用，这是个问题
        def fn(current_list, index):
            if index == len(nums):
                ans.append(current_list.copy())
                return
            # 用
            current_list.append(nums[index])
            fn(current_list, index+1)
            # 回溯
            current_list.pop()
            # 不用
            fn(current_list, index+1)
        
        fn([], 0)
        
        return ans
````
