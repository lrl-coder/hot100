# 131. 分割回文串

给你一个字符串 s，请你将 s 分割成一些 子串，使每个子串都是 回文串 。返回 s 所有可能的分割方案。

示例 1：
```text
输入：s = "aab"
输出：[["a","a","b"],["aa","b"]]
```
示例 2：
```text
输入：s = "a"
输出：[["a"]]
```

````python
class Solution:
    def partition(self, s: str) -> List[List[str]]:
        result = []
        def backtrack(path, start):
            nonlocal result
            if start == len(s):
                result.append(path.copy())
                return
            # 循环所有的分割位置
            for i in range(start, len(s)):
                sub_s = s[start:i+1]
                if sub_s == sub_s[::-1]:
                    path.append(sub_s)
                    backtrack(path, i+1)
                    path.pop()

        backtrack([], 0)
        return result
````
