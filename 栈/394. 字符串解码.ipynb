# 394. 字符串解码

给定一个经过编码的字符串，返回它解码后的字符串。

编码规则为: k[encoded_string]，表示其中方括号内部的 encoded_string 正好重复 k 次。注意 k 保证为正整数。

你可以认为输入字符串总是有效的；输入字符串中没有额外的空格，且输入的方括号总是符合格式要求的。

此外，你可以认为原始数据不包含数字，所有的数字只表示重复的次数 k ，例如不会出现像 3a 或 2[4] 的输入。

测试用例保证输出的长度不会超过 $10^5$。

示例 1：
```text
输入：s = "3[a]2[bc]"
输出："aaabcbc"
```
示例 2：
```text
输入：s = "3[a2[c]]"
输出："accaccacc"
```
示例 3：
```text
输入：s = "2[abc]3[cd]ef"
输出："abcabccdcdcdef"
```
示例 4：
```text
输入：s = "abc3[cd]xyz"
输出："abccdcdcdxyz"
```

## 解题思路

这道题的终极奥义，其实就是四个字：保护现场。

以 `a3[b2[c]]` 为例，想象你在打游戏（当前积累的字符串 res = "a"）。突然，你接到了一个支线任务 3[。
为了去打支线任务，你必须把你现在打出的进度（"a"）以及支线任务的奖励倍数（3），统统存到一个保险箱（栈）里，然后清空双手，去括号里面重新打怪。
打完支线任务遇到 ]，你再把保险箱里的老进度拿出来，和新打出来的战利品拼在一起。

我们需要两个临时变量，和一个装元组的保险箱（栈）：

- multi：当前正在计算的倍数。
- res：当前正在拼接的字符串。
- stack：用来保存遇到 [ 之前的 (res, multi) 快照。

````python
class Solution:
    def decodeString(self, s: str) -> str:
        stack = []   # 保险箱：用来存储 (历史res, 当前multi)
        res = ""     # 当前正在手里的字符串
        multi = 0    # 当前正在手里的倍数

        for char in s:
            if char.isdigit():
                # 遇到数字，可能有多位数，所以要 *10 累加
                multi = multi * 10 + int(char)
            elif char == '[':
                # 遇到左括号，准备进入里层世界！
                # 必须把外层的 res 和 multi 锁进保险箱
                stack.append((res, multi))
                # 清空双手，准备处理括号里面的新内容
                res = ""
                multi = 0
            elif char == ']':
                # 遇到右括号，里层世界打完了！
                # 从保险箱里拿出外层的进度 (last_res) 和倍数 (cur_multi)
                last_res, cur_multi = stack.pop()
                # 拼接：历史进度 + 倍数 * 里层打出的战利品
                res = last_res + cur_multi * res
            else:
                # 遇到普通字母，直接拿到手里
                res += char

        return res
````
