# 155. 最小栈

设计一个支持 push ，pop ，top 操作，并能在常数时间内检索到最小元素的栈。

实现 MinStack 类:

- MinStack() 初始化堆栈对象。
- void push(int value) 将元素 value 推入堆栈。
- void pop() 删除堆栈顶部的元素。
- int top() 获取堆栈顶部的元素。
- int getMin() 获取堆栈中的最小元素。
  
示例 1:
```text
输入：
["MinStack","push","push","push","getMin","pop","top","getMin"]
[[],[-2],[0],[-3],[],[],[],[]]

输出：
[null,null,null,null,-3,null,0,-2]

解释：
MinStack minStack = new MinStack();
minStack.push(-2);
minStack.push(0);
minStack.push(-3);
minStack.getMin();   --> 返回 -3.
minStack.pop();
minStack.top();      --> 返回 0.
minStack.getMin();   --> 返回 -2.
```

## 思路

只需要记录当前栈顶元素和当前最小元素。

我们需要让栈里的每一个元素，都记住“当它作为栈顶时，栈里的最小值是多少”。直接压入元组 (当前值, 当前的历史最小值)。这就好比给每个数字拍了一张“快照”。

很像 git！

````python
class MinStack:
    from collections import deque
    def __init__(self):
        self.q = deque()


    def push(self, value: int) -> None:
        if not self.q:
            self.q.append((value, value))
        else:
            min_val = min(self.q[-1][1], value)
            self.q.append((value, min_val))        

    def pop(self) -> None:
        x = self.q.pop()
        return x[0]        

    def top(self) -> int:
        return self.q[-1][0]

    def getMin(self) -> int:
        return self.q[-1][1]        


# Your MinStack object will be instantiated and called as such:
# obj = MinStack()
# obj.push(value)
# obj.pop()
# param_3 = obj.top()
# param_4 = obj.getMin()
````
