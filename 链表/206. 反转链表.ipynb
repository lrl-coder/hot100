# 反转链表

给你单链表的头节点 head ，请你反转链表，并返回反转后的链表。





## 把每一个箭头都反转



这个题本身不难，直接把所有的箭头方向都反向即可，两个步骤：

也就是 p->next = p->front.

需要特别注意的是要存储指针的值，防止断链，并且要注意赋值的副作用！


```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        # 把箭头的方向都反向即可
        if not head:
            return None
        p = head
        front = None
        while p:
            t = p.next
            p.next = front
            front = p
            p = t
        return front      
```
