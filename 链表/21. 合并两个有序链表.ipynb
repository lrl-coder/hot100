# 21. 合并两个有序链表

将两个升序链表合并为一个新的 升序 链表并返回。新链表是通过拼接给定的两个链表的所有节点组成的。 

输入：l1 = [1,2,4], l2 = [1,3,4]

输出：[1,1,2,3,4,4]

解题思路和合并两个有序列表差不多。

下面给出代码：

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:       
        p = list1
        q = list2
        ans = ListNode()
        current = ans
        while p and q:
            if p.val < q.val:
                current.next = p
                current = p
                p = p.next
            else:
                current.next = q
                current = q
                q = q.next
        if p:
            current.next = p
        if q:
            current.next = q
        
        return ans.next
```
