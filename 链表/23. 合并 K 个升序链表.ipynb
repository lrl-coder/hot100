# 23. 合并 K 个升序链表

给你一个链表数组，每个链表都已经按升序排列。

请你将所有链表合并到一个升序链表中，返回合并后的链表。

## 邪修

把所有链表的值全取出来，放进数组里排序，然后再重新组装成一条新的链表。

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        arr = []
        for head in lists:
            p = head
            while p:
                arr.append(p.val)
                p = p.next

        if not arr:
            return None

        arr.sort()
        print(arr)

        head = ListNode(next=None)
        p = head
        for i in arr:
            print(i)   
            node = ListNode(val=i)
            p.next = node
            p = p.next
        return head.next
```

## 优先队列（小根堆）

python 的 heapq 默认就是最小堆，他会自动帮我们维护：无论怎么插入数据，每次弹出的永远是当前集合中最小的那个元素。

使用最小堆，堆里最多只需要维持 $k$ 个元素（每个链表当前的头节点）。每次弹出一个最小的，然后把被弹出节点的下一个节点塞进堆里（如果有的话），循环往复。直到堆中没有元素。

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

import heapq
class Solution:
    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        print(help(heapq))
        heap = []

        for i,item in enumerate(lists):
            if item:
                heapq.heappush(heap, (item.val, i, item))

        head = ListNode(next=None)
        p = head
        while heap:
            val, i, min_node = heapq.heappop(heap)
            p.next = min_node
            p = p.next
            if min_node and min_node.next:
                heapq.heappush(heap, (min_node.next.val, i, min_node.next))

        return head.next
```
