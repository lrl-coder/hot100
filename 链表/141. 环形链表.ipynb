# 141. 环形链表

给你一个链表的头节点 head ，判断链表中是否有环。

如果链表中有某个节点，可以通过连续跟踪 next 指针再次到达，则链表中存在环。 为了表示给定链表中的环，评测系统内部使用整数 pos 来表示链表尾连接到链表中的位置（索引从 0 开始）。注意：pos 不作为参数进行传递 。仅仅是为了标识链表的实际情况。

如果链表中存在环 ，则返回 true 。 否则，返回 false 。

## 辅助数组

开一个辅助 set 用于记录节点，如果有重复的就说明有环，反之无环。


```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def hasCycle(self, head: Optional[ListNode]) -> bool:
        nodes = set()
        p = head
        while p:
            if p in nodes:
                return True
            nodes.add(p)
            p = p.next
        return False        
```

## 优化：快慢指针法

把链表想象成一条跑道，让两个指针同时从起点起跑。

- 慢指针每次跑 1 步
- 快指针每次跑 2 步

1. 如果链表没有环，那么快指针先跑到终点（None）
2. 如果链表有环，那么快指针一定会追上慢指针，两指针相遇。
