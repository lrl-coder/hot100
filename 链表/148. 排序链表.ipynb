## 148. 排序链表

给你链表的头结点 head ，请将其按 升序 排列并返回 排序后的链表 。



## 邪修

用一个数组存储所有值，sort 排序，再填回链表。投机取巧法。

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def sortList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        arr = []
        p = head
        while p:
            arr.append(p.val)
            p = p.next
        p = head
        i = 0
        arr.sort()
        while p:
            p.val = arr[i]
            p = p.next
            i += 1
        return head
```
