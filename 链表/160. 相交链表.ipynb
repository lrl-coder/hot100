# 相交链表

给你两个单链表的头节点 headA 和 headB ，请你找出并返回两个单链表相交的起始节点。如果两个链表不存在相交节点，返回 null 。

例如图示两个链表在节点 c1 开始相交：

![相交的链表](https://assets.leetcode.cn/aliyun-lc-upload/uploads/2018/12/14/160_statement.png)

## 暴力解法：存节点

用 set 存储第一条链的节点，然后遍历第二条链，看有没有节点在集合中，如果有，该节点就是相交点，否则两条链不相交。

**注意**：一定要用 set ，因为他底层用的是 哈希实现，查找元素是否存在的时间效率为 O1，用 list 会超时！

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> Optional[ListNode]:
        nodes = set()
        pA, pB = headA, headB
        while pA:
            nodes.add(pA)
            pA = pA.next
        while pB:
            if pB in nodes:
                return pB
            pB = pB.next
        return None        
```

## 优化：双指针



核心就是利用了 (x + z) + y = (y + z) + x.

> 当我在我的路上走过一遍没有遇见你时，那么我会接着来到你走过的路走一遍，如果我们心有灵犀，那么我们会遇见。如果确实没有缘分，那么即使再走一遍也不会相遇。

相当于分两种情况讨论：

1. x 等于 y，那么第一次就会遇见
2. x 不等于 y，那么走第二遍时最终的结果是两个指针都指向同一个位置，但是分两种情况：
   1. 相遇：指向的位置即为相交点
   2. 不相遇：指向的位置为 None

根据题目意思，我们直接返回指针即可。
