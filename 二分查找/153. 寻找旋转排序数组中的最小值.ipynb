# 153. 寻找旋转排序数组中的最小值

已知一个长度为 n 的数组，预先按照升序排列，经由 1 到 n 次 旋转 后，得到输入数组。例如，原数组 nums = [0,1,2,4,5,6,7] 在变化后可能得到：
- 若旋转 4 次，则可以得到 [4,5,6,7,0,1,2]
- 若旋转 7 次，则可以得到 [0,1,2,4,5,6,7]
  
注意，数组 [a[0], a[1], a[2], ..., a[n-1]] 旋转一次 的结果为数组 [a[n-1], a[0], a[1], a[2], ..., a[n-2]] 。

给你一个元素值 互不相同 的数组 nums ，它原来是一个升序排列的数组，并按上述情形进行了多次旋转。请你找出并返回数组中的 最小元素 。

你必须设计一个时间复杂度为 O(log n) 的算法解决此问题。

```python
class Solution:
    def findMin(self, nums: List[int]) -> int:
        # 一刀切两半，必定有一半是有顺序的
        ans = float('inf')
        left = 0
        right = len(nums)-1
        while left <= right:
            mid = (left+right)//2
            # 左半有序
            if nums[left] <= nums[mid]:
                # 左半部分的最小值
                ans = min(ans, nums[left])
                # 再看右边
                left = mid+1
            else:
                ans = min(ans, nums[mid])
                right=mid-1
        return ans
        
```

## 优化

在这个旋转数组的折线图里，必定分为两段：左边的“高地” 和 右边的“低谷”。而我们要找的最小值，就是低谷的第一个元素。

在二分查找中，我们需要一个“标尺”来判断当前砍下的一刀（mid）到底落在高地还是低谷。
这道题最核心的秘诀在于：永远拿 nums[mid] 去和 nums[right] 比！

当你算出 mid 后，只有两种情况：

1. nums[mid] > nums[right]：
   这说明 mid 当前站在“左边的高地”上。既然在高地上，那么真正的最小值（坑底）必定在它的右边。
   → 搜索区间毫不犹豫地向右压缩：left = mid + 1。

2. nums[mid] < nums[right]：
   这说明 mid 已经掉进了“右边的低谷”里。在这个低谷里，它是绝对有序的。既然 mid 处于低谷，那么它本身有可能就是最小值，或者最小值在它的左边。
   → 搜索区间向左压缩，但必须保留 mid：right = mid。
