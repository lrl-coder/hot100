# 34. 在排序数组中查找元素的第一个和最后一个位置

给你一个按照非递减顺序排列的整数数组 nums，和一个目标值 target。请你找出给定目标值在数组中的开始位置和结束位置。

如果数组中不存在目标值 target，返回 [-1, -1]。

你必须设计并实现时间复杂度为 O(log n) 的算法解决此问题。

示例 1：
```text
输入：nums = [5,7,7,8,8,10], target = 8
输出：[3,4]
```
示例 2：
```text
输入：nums = [5,7,7,8,8,10], target = 6
输出：[-1,-1]
```
示例 3：
```text
输入：nums = [], target = 0
输出：[-1,-1]
```

````python
class Solution:
    def searchRange(self, nums: List[int], target: int) -> List[int]:
        # 查找第一个大于等于x的位置
        # 蓝区间：小于x
        # 红区间：大于等于x
        def searchPosition(nums, x):
            left, right = 0, len(nums)-1
            while left <= right:
                mid = (left+right)//2
                if nums[mid] < x:
                    left = mid+1
                elif nums[mid] >= x:
                    right = mid-1
            return left

        start = searchPosition(nums, target)
        if start >= len(nums) or nums[start]!=target:
            return [-1, -1]
        end = searchPosition(nums, target+1) - 1
        
        return [start, end]
````
