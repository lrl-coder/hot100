# 无重复字符的最长子串

给定一个字符串 s ，请你找出其中不含有重复字符的 最长 子串 的长度。

示例 1:

输入: s = "abcabcbb"

输出: 3 

解释: 因为无重复字符的最长子串是 "abc"，所以其长度为 3。注意 "bca" 和 "cab" 也是正确答案。

示例 2:

输入: s = "bbbbb"

输出: 1

解释: 因为无重复字符的最长子串是 "b"，所以其长度为 1。

示例 3:

输入: s = "pwwkew"

输出: 3

解释: 因为无重复字符的最长子串是 "wke"，所以其长度为 3。
     请注意，你的答案必须是 子串 的长度，"pwke" 是一个子序列，不是子串。
 

提示：

0 <= s.length <= 5 * 104

s 由英文字母、数字、符号和空格组成

## 朴素版滑动窗口
用left、right指针去维护一个不含重复字符的子串，right不断向右滑动，若新的字符与现有窗口字符串有重复的，左指针需要右移去保证窗口没有重复字符。

这样时间复杂度是 n^2， 空间复杂度 1。

```python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        if not len(s):
            return 0
        # 滑动窗口的左右端点
        left, right = 0, 0
        ans = 1
        while right < len(s) - 1:
            # 检查在滑动窗口的字符串与右边的字符是否有重复
            ch = s[right + 1]
            idx = -1
            for i in range(right, left - 1, -1):
                if ch == s[i]:
                    idx = i
                    break
            if idx >= 0:
                left = idx + 1
            right += 1
            ans = max(ans, right - left + 1) 
        return ans
```
