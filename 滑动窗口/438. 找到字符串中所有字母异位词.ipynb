# 找到字符串中所有字母异位词

给定两个字符串 s 和 p，找到 s 中所有 p 的 异位词 的子串，返回这些子串的起始索引。不考虑答案输出的顺序。


示例 1:

输入: s = "cbaebabacd", p = "abc"

输出: [0,6]

解释:

起始索引等于 0 的子串是 "cba", 它是 "abc" 的异位词。

起始索引等于 6 的子串是 "bac", 它是 "abc" 的异位词。

 示例 2:

输入: s = "abab", p = "ab"

输出: [0,1,2]

解释:

起始索引等于 0 的子串是 "ab", 它是 "ab" 的异位词。

起始索引等于 1 的子串是 "ba", 它是 "ab" 的异位词。

起始索引等于 2 的子串是 "ab", 它是 "ab" 的异位词。 

提示:

1 <= s.length, p.length <= 3 * 10^4

s 和 p 仅包含小写字母

## 暴力解法：排序+遍历

一看见这个题目，我的第一直觉就是先把字符串 p 从小到大排序，然后再遍历 s ，收集结果。

能通过，但是运行时间较长。

```python
class Solution:
    def findAnagrams(self, s: str, p: str) -> List[int]:
        len_s, len_p = len(s), len(p)
        p = ''.join(sorted(p))
        ans = []
        for i in range(len_s - len_p + 1):
            sub_s = ''.join(sorted(s[i: i + len_p]))
            if sub_s == p:
                ans.append(i)
        return ans
        
```

## 滑动窗口解法

进一步观察，题目中写了：s 和 p 仅包含小写字母。

这就给了我们一个很强的提示：最多出现26中字符，可以采用空间换时间的思想了！

```python
class Solution:
    def findAnagrams(self, s: str, p: str) -> list[int]:
        len_s, len_p = len(s), len(p)
        if len_s < len_p:
            return []
            
        # 用长度为 26 的数组代替字典
        count_s = [0] * 26
        count_p = [0] * 26
        
        # 初始化第一个窗口和 p 的词频
        for i in range(len_p):
            count_s[ord(s[i]) - 97] += 1  # 97 是 'a' 的 ASCII 码，也可以写 ord('a')
            count_p[ord(p[i]) - 97] += 1
            
        ans = [0] if count_s == count_p else []
        
        # 开始滑动窗口
        for i in range(len_p, len_s):
            # 右侧新字符进入窗口
            count_s[ord(s[i]) - 97] += 1
            # 左侧老字符移出窗口
            count_s[ord(s[i - len_p]) - 97] -= 1
            
            # 直接比较两个数组
            if count_s == count_p:
                ans.append(i - len_p + 1)
                
        return ans
```
