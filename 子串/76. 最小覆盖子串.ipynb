76. 最小覆盖子串

给定两个字符串 s 和 t，长度分别是 m 和 n，返回 s 中的 最短窗口 子串，使得该子串包含 t 中的每一个字符（包括重复字符）。如果没有这样的子串，返回空字符串 ""。

测试用例保证答案唯一。


示例 1：


输入：s = "ADOBECODEBANC", t = "ABC"

输出："BANC"

解释：最小覆盖子串 "BANC" 包含来自字符串 t 的 'A'、'B' 和 'C'。

示例 2：

输入：s = "a", t = "a"

输出："a"

解释：整个字符串 s 是最小覆盖子串。

示例 3:

输入: s = "a", t = "aa"

输出: ""

解释: t 中两个字符 'a' 均应包含在 s 的子串中，
因此没有符合条件的子字符串，返回空字符串。

提示：

m == s.length

n == t.length

1 <= m, n <= 10^5

s 和 t 由英文字母组成

## 求解思路

首先来看覆盖的含义：s的子串每个字母的出现次数，都大于等于t中每个字母出现的次数，这个就叫做覆盖。

**总体思路**：贪心吃撑，再吐出多余的。

想象你在餐厅吃自助餐（遍历字符串 s），你的目标是吃够特定的营养（字符串 t 里的所有字符）。进食流程：

1. 进食（右指针右移）：疯狂往嘴里塞东西（扩大窗口），指导你吃够了所有需要的营养，此时，得到了一个覆盖子串，但这可能不是最小的，因为你可能吃了很多不相关的垃圾食品，或者同一种营养吃多了。
2. 吐出（左指针右移）：既然已经吃够了，为了追求最小，我们开始从最前面把吃进去的东西吐出来（缩小窗口）。
   1. 如果吐出来之后营养依旧达标，那就继续吐！此时窗口不断缩小，不断更新最小长度。
   2. 直到营养不达标为止。
3. 循环往复，直至右指针走到尽头。

```python
import collections
class Solution:
    def minWindow(self, s: str, t: str) -> str:
        cnt_s = Counter()
        cnt_t = Counter(t)
        left, right = 0, 0
        ans = []
        while right < len(s):
            cnt_s[s[right]] += 1
            while cnt_s >= cnt_t:
                cnt_s[s[left]] -= 1
                if not ans or right - left + 1 < ans[1] - ans[0] + 1:
                    ans = [left, right]
                left += 1
            right += 1
        return s[ans[0]:ans[1] + 1] if ans else ''
```
